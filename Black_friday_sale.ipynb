***#MODEL BUILDING***

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.metrics import accuracy_score
```

```python
bfs_df.columns
```

```python
def purchase_range(purchase_amount):
  if purchase_amount<=1000:
    pur='Less than 1000'
  elif purchase_amount>1000 and purchase_amount<=5000:
    pur='Less than 5000'
  elif purchase_amount>5000 and purchase_amount<=10000:
    pur='Less than 10000'
  elif purchase_amount>10000 and purchase_amount<=15000:
    pur='Less than 15000'
  else:
    pur='Greater than 15000'
  return pur
```

```python
bfs_df['Purchase_Category']= bfs_df['Purchase_amount'].apply(purchase_range)
```

```python
bfs_df.dtypes
```

```python
bfs_df= bfs_df[['Gender', 'Occupation', 'City_Category','Stay_In_Current_City_Years', 'Marital_Status', 'Avg_Age', 'Purchase_Category']]
```

```python
bfs_df
```

```python
x= bfs_df[[ 'Gender', 'Occupation', 'City_Category','Stay_In_Current_City_Years', 'Marital_Status', 'Avg_Age']]
y= bfs_df[['Purchase_Category']]
```

```python
x
```

```python
y
```

```python
le=LabelEncoder()
x['Gender_n']=le.fit_transform(x['Gender'])
x['City_Catefory_n']=le.fit_transform(x['City_Category'])
x['Stay_In_Current_City_Years_n']=le.fit_transform(x['Stay_In_Current_City_Years'])
```

```python
x.head()
```

```python
y['Cat_n']=le.fit_transform(y['Purchase_Category'])
y.head()
```

```python
y=y['Cat_n']
```

```python
x=x[['Gender_n', 'Occupation', 'City_Category_n','Stay_In_Current_City_Years_n', 'Marital_Status', 'Avg_Age']]
x.head()
```

```python
x.dtypes
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x_test=sc.transform(x)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr=LogisticRegression(max_iter=1000).fit(x_train,y_train)
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
```

```python
y_pred_lr= lr.predict(x_test)
y_pred_dt= lr.predict(x_test)
y_pred_rf= lr.predict(x_test)
y_pred_knn= lr.predict(x_test)
```

```python
lr_score=round((accuracy_score(y_pred_lr,y_test)*100))
dt_score=round((accuracy_score(y_pred_dt,y_test)*100))
rf_score=round((accuracy_score(y_pred_rf,y_test)*100))
knn_score=round((accuracy_score(y_pred_knn,y_test)*100))

print("The Accuracy score for Logistic Regression is ", lr_score, "%")
print("The Accuracy score for Decision Tree is ", dt_score, "%")
print("The Accuracy score for Random Forrest is ", rf_score, "%")
print("The Accuracy score for KNN Score is ", knn_score, "%")
```
